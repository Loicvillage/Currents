```python
import pandas
pandas.set_option("display.max_columns", None)
import numpy as np
import copy
from urllib.error import HTTPError
import matplotlib.pyplot as plt
import tqdm
import sklearn.linear_model
import sklearn.preprocessing 
import sklearn.model_selection
import sklearn.metrics

import seaborn as sns
```

```python
startyear = 2004
endyear = 2014
```

***
# HOME TOV

```python
TRTR = pandas.read_csv("S_Full_HoAw_" + str(startyear) + "_" + str(endyear) + ".csv")
TRTR = TRTR.loc[TRTR['GameNum'] >= 25].dropna()

coefs_b1 = []
coefs_b2 = []
for year in range(startyear, endyear + 1):
    print (year)
    TFG = copy.deepcopy(TRTR.loc[TRTR['Season'] == year])
    home_feats = TFG.loc[:,['home.GG.avg.prob_TOV', 'away.BG.avg.prob_TOV']]
    home_tar = TFG.loc[:,['home.GG.curr.prob_TOV']]
    home_Model = sklearn.linear_model.LinearRegression(fit_intercept = False).fit(home_feats, home_tar )
    TFG['pred'] = home_Model.predict(home_feats)
    print("HOME_________")
    #print(home_Model.coef_)
    coefs_b1 = coefs_b1 + [home_Model.coef_[0][0]]
    coefs_b2 = coefs_b2 + [home_Model.coef_[0][1]]
    print(sum(sum(home_Model.coef_)))
    #print ("F1 R2: " + str(np.round(sklearn.metrics.r2_score(home_tar,TFG.loc[:,['home.GG.avg.prob_TOV']]),5)))
    #print ("F2 R2: " + str((np.round(sklearn.metrics.r2_score(home_tar,TFG.loc[:,['away.BG.avg.prob_TOV']]),5))))
    print ("Yhat R2: " + str((np.round(sklearn.metrics.r2_score(home_tar, TFG['pred']),5))))
    print ("Yhat MedAE: " + str((np.round(sklearn.metrics.median_absolute_error(home_tar, TFG['pred']),5))))
    print ("Yhat MeanAE: " + str((np.round(sklearn.metrics.mean_absolute_error(home_tar, TFG['pred']),5))))
print("Coefs b1")
print(np.round(coefs_b1,5))
print("Sum Coefs b1:" + str(np.round(np.sum(coefs_b1),5)))
print("Avg Coefs b1:" + str(np.round(np.sum(coefs_b1) / (endyear - startyear + 1),5)))
print("Coefs b2")
print(np.round(coefs_b2,5))
print("Sum Coefs b2:" + str(np.round(sum(coefs_b2),5)))
print("Avg Coefs b2:" + str(np.round(np.sum(coefs_b2) / (endyear - startyear + 1),5)))
print("Avg b1 + b2 Coefs:" + str(np.round((np.sum(coefs_b2) + np.sum(coefs_b1)) / ((endyear - startyear + 1)),5)))

TFG = copy.deepcopy(TRTR)#.loc[TRTR['Season'] == year])
home_feats = TFG.loc[:,['home.GG.avg.prob_TOV', 'away.BG.avg.prob_TOV']]
home_tar = TFG.loc[:,['home.GG.curr.prob_TOV']]
home_Model = sklearn.linear_model.LinearRegression(fit_intercept = False).fit(home_feats, home_tar )
TFG['pred'] = home_Model.predict(home_feats)
print("ALL_________")
print(home_Model.coef_)
# coefs_b1 = coefs_b1 + [home_Model.coef_[0][0]]
# coefs_b2 = coefs_b2 + [home_Model.coef_[0][1]]
print(sum(sum(home_Model.coef_)))
#print ("F1 R2: " + str(np.round(sklearn.metrics.r2_score(home_tar,TFG.loc[:,['home.GG.avg.prob_TOV']]),5)))
#print ("F2 R2: " + str((np.round(sklearn.metrics.r2_score(home_tar,TFG.loc[:,['away.BG.avg.prob_TOV']]),5))))
print ("Yhat R2: " + str((np.round(sklearn.metrics.r2_score(home_tar, TFG['pred']),5))))
print ("Yhat MedAE: " + str((np.round(sklearn.metrics.median_absolute_error(home_tar, TFG['pred']),5))))
print ("Yhat MeanAE: " + str((np.round(sklearn.metrics.mean_absolute_error(home_tar, TFG['pred']),5))))
```
